```python
N = 50
A = np.diag(np.repeat(2,N)) + np.diag(np.repeat(-1,N-1),1) + np.diag(np.repeat(-1,N-1),-1)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(A,cmap='binary')
plt.axis(False)
plt.savefig('../img/cover.png',dpi=300,bbox_inches='tight',pad_inches=0.0)
plt.show()
```
